Three 'Seaborn' functions that Cover(almost) all your visualization tasks:
- Relplot: Used for creating relational plots.
- Displot: Used for creating distributions plots.
- Catplot: Used for creating categorical plots.

### __Plotting categorical data__
- `1) Categorical scatter plots` (atleast one variable is categorical)
  - Strip plot
  - Swarm plot
- `2) Categorical distribution plots`
  - Box plot
  - Voilin plot
- `3) Categorical estimate plots`
  - Point plot
  - Bar plot
  - Countplot
- `4) Categorical categorical plot`
  - Heat maps
  - Cluster maps
- `5) Potting side by side graphs using Facetgrid`

#### __Categorical scatter plots__
- x axis: categorical data
- y axis: numeric data

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

#### __`Strip plot`__

```python
sns.catplot(x='day', y='tip', kind='strip', data=tips) # by default it is kind = 'strip'
plt.show()
```

```python
# There is is algorithm called jitter works behind it, which adds some noise in the data and that's why the points do not overlap.
sns.catplot(x='day', y='tip', kind='strip',jitter=2, data=tips)
plt.show()
```

#### __`Swarmplot`__
- gives you additional information or we can say that it also shows the distribution of the data

```python
sns.catplot(x='day', y='tip', kind='swarm', data=tips)
plt.show()
```

```python
# 'hue' semantic parameter
sns.catplot(x='day', y='tip', kind='swarm', hue='sex', data=tips)
plt.show()
```

```python
# shortcut
sns.swarmplot(x='day', y='tip',data=tips)
plt.show()
```
