### __`Pair plots`__
Pair plots are used to plot multiple pairwise bivariate distributions in dataset. This creates a matrix of axes and shows the relationship for each pair of column in a dataframe. By default, it also drws the univariate distribution of each variable on the diagonal axes.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
# it shows histogram since it will be a univariate analysis and scatter plot can not be used for univariate analysis.
sns.pairplot(iris)
plt.show()
```

```python
# using 'hue' as a semantic parameter
sns.pairplot(iris,hue='species')
plt.show()
```
