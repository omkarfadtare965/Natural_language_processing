```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
df1 = pd.read_csv(
    r'C:\Users\MY\OneDrive\Desktop\Visualization\Plotly\matches.csv')
df2 = pd.read_csv(
    r'C:\Users\MY\OneDrive\Desktop\Visualization\Plotly\deliveries.csv')

df = df2.merge(df1, left_on='match_id', right_on='id')
df.head(3)
```

```python
top50 = df.groupby('batsman')['batsman_runs'].sum().sort_values(
    ascending=False).head(50).index.tolist()
new_ipl = df[df['batsman'].isin(top50)]
```

```python
# Calculating SR
# SR=[(number of runs scored)/(number of balls played)]*100
runs = new_ipl.groupby('batsman')['batsman_runs'].sum()
balls = new_ipl.groupby('batsman')['batsman_runs'].count()

sr = (runs/balls)*100

sr = sr.reset_index()
sr
```

```python
# Calculating Avg
# Avg=(Total number of Runs)/(Number of outs)

# Calculating number of outs for top 50 batsman
out = df[df['player_dismissed'].isin(top50)]

nouts = out['player_dismissed'].value_counts()

avg = runs/nouts

avg = avg.reset_index()
avg.rename(columns={'index': 'batsman', 0: 'avg'}, inplace=True)

avg = avg.merge(sr, on='batsman')
avg
```

### __1] Scatter plot__

```python
from matplotlib.pyplot import title
```

```python
trace = go.Scatter(x=avg['avg'], y=avg['batsman_runs'], mode='markers',
                   text=avg['batsman'], marker={'color': '#00a65a', 'size': 16})
data = [trace]
layout = go.Layout(title='Batsman avg vs SR', xaxis={
                   'title': 'Batsman avg'}, yaxis={'title': 'Batsman SR'})
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
pyo.plot(fig, filename='scatter.html')
```

```python
# Multiple Line Charts

def batsman_comp(*name):
    data=[]
    for i in name:
        single=df[df['batsman']==i]
        performance=single.groupby('season')['batsman_runs'].sum().reset_index()

        trace=go.Scatter(x=performance['season'],y=performance['batsman_runs']
                         ,mode='lines + markers',name=i)
        
        data.append(trace)
    
    layout=go.Layout(title='Batsman Record Comparator',
                xaxis={'title':'Season'},
                yaxis={'title':'Runs'})

    fig=go.Figure(data=data,layout=layout)

    pyo.plot(fig,filename='year_by_year_line3')
```

### __3] Bar plot__
- Used to show relation between one categorical and one numeric variable
- There are three types of bar graph
   - Nested bar graph
   - Stacked bar graph
   - Overlayed bar graph

```python
top10=df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10).index.tolist()
top10_df=df[df['batsman'].isin(top10)]
```

```python
top10_score = top10_df.groupby('batsman')['batsman_runs'].sum().reset_index()
top10_score
```

```python
trace = go.Bar(x= top10_score['batsman'], y=top10_score['batsman_runs'])
data = [trace]
layout = go.Layout(title = 'Top 10 IPL Batsman', xaxis={'title':'Batsman'},yaxis={'title':'Total runs'})
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='bar_plot.html')
```

```python
iw = top10_df.groupby(['batsman', 'inning'])[
    'batsman_runs'].sum().reset_index()
mask = iw['inning'] == 1
mask2 = iw['inning'] == 2
one = iw[mask]
two = iw[mask2]


one.rename(columns={'batsman_runs': '1st Innings'}, inplace=True)
two.rename(columns={'batsman_runs': '2nd Innings'}, inplace=True)

final = one.merge(two, on='batsman')[['batsman', '1st Innings', '2nd Innings']]

final
```

```python
# Overlay

trace1 = go.Bar(x=final['batsman'], y=final['1st Innings'], name='1st Innings',
                marker={'color': '#a6a65a'})
trace2 = go.Bar(x=final['batsman'], y=final['2nd Innings'], name='2nd Innings',
                marker={'color': '#00a65a'})

data = [trace1, trace2]

layout = go.Layout(title='Inning wise scores',
                   xaxis={'title': 'Batsman'},
                   yaxis={'title': 'Runs'},
                   barmode='overlay')

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='overlay_bar.html')
```

```python
# Stacked

trace1 = go.Bar(x=final['batsman'], y=final['1st Innings'], name='1st Innings',
                marker={'color': '#a6a65a'})
trace2 = go.Bar(x=final['batsman'], y=final['2nd Innings'], name='2nd Innings',
                marker={'color': '#00a65a'})

data = [trace1, trace2]

layout = go.Layout(title='Inning wise scores',
                   xaxis={'title': 'Batsman'},
                   yaxis={'title': 'Runs'},
                   barmode='stack')

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='stacked_bar.html')
```

```python
# Nested

trace1 = go.Bar(x=final['batsman'], y=final['1st Innings'], name='1st Innings',
                marker={'color': '#a6a65a'})
trace2 = go.Bar(x=final['batsman'], y=final['2nd Innings'], name='2nd Innings',
                marker={'color': '#00a65a'})

data = [trace1, trace2]

layout = go.Layout(title='Inning wise scores',
                   xaxis={'title': 'Batsman'},
                   yaxis={'title': 'Runs'})

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='nested_bar.html')
```

### __5] Box plot__

### __6] Histogram__

```python
x = df2.groupby('batsman')['batsman_runs'].count() > 150
x = x[x].index.tolist()

new = df2[df2['batsman'].isin(x)]

runs = new.groupby('batsman')['batsman_runs'].sum()
balls = new.groupby('batsman')['batsman_runs'].count()

sr = (runs/balls)*100
sr = sr.reset_index()
sr
```

```python
trace = go.Histogram(x=sr['batsman_runs'])
data = [trace]
layout = go.Layout(title='strike rate analysis',
                   xaxis={'title': 'strike rates'})

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='hist1.html')
```

```python
trace = go.Histogram(x=sr['batsman_runs'], xbins={'size': 2})
data = [trace]
layout = go.Layout(title='strike rate analysis',
                   xaxis={'title': 'strike rates'})

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='hist2.html')
```

```python
trace = go.Histogram(x=sr['batsman_runs'], xbins={
                     'size': 2, 'start': 50, 'end': 100})
data = [trace]
layout = go.Layout(title='strike rate analysis',
                   xaxis={'title': 'strike rates'})

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='hist3.html')
```

### __8] Heatmap__
- It can be plotted on 3 variables, 2 can be categorical or numeric 

```python
six = df2[df2['batsman_runs'] == 6]
six = six.groupby(['batting_team', 'over'])[
    'batsman_runs'].count().reset_index()
six
```

```python
trace = go.Heatmap(x=six['batting_team'], y=six['over'], z=six['batsman_runs'])

data = [trace]

layout = go.Layout(title='Six Heatmap')
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig,filename='heat1.html')
```

```python
from plotly import tools

# Side by Side Heatmap
dots = df2[df2['batsman_runs'] == 0]
dots = dots.groupby(['batting_team', 'over'])[
    'batsman_runs'].count().reset_index()

trace1 = go.Heatmap(x=six['batting_team'], y=six['over'],
                    z=six['batsman_runs'].values.tolist())

trace2 = go.Heatmap(x=dots['batting_team'], y=dots['over'],
                    z=dots['batsman_runs'].values.tolist())

fig = tools.make_subplots(rows=1, cols=2, subplot_titles=[
                          "6's", "0's"], shared_yaxes=True)

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)

pyo.plot(fig,filename='heat2.html')
```
