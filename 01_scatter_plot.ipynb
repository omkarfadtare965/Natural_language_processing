### __Relational/Scatter plots__
Relational or statistical plots are used to perform multivariate analysis
- Scatter plots
- Line plots (specifically used for time series data)
- FacetGrids plot (plotting multiple graphs side by side)

#### __`Scatter plots`__
- The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer substantial amount of information about, whether there is any meaningful relationship between them.
- scatter plots can be used for both categorical as well as numeric data, but its more commonly used to depict relationship between numerical data.
- There are two ways to draw a scatter plot in seaborn
   - using `relplot` (figure level function)
   - using `scatterplot` (axes level function)
- As we will see, these visualization can be quite helpful because they use simple and easily understandable representation of data even on complex data by mapping up to three additional variables using semantics of `'hue'`, `'size'`, `'style'`.
- You can use either scatter plot or relplot, both are the same.
- `'hue'`: categorical data, `'style'`: categorical data, `'size'`: continuous or numeric data.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head(5)
```

```python
# plotting a scatter plot between total_bill and tip
# Using relplot
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data) # by default kind = 'scatter'
plt.show()
```

```python
# 'hue' semantic parameter
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
plt.show()
```

```python
# 'style' semantic parameter
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex', data=data)
plt.show()
```

```python
# 'size' semantic parameter
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex',size='size', data=data)
plt.show()
```

```python
# To change the size 
sns.relplot(x='total_bill', y='tip', hue='smoker',
            style='sex', size='size',sizes=(15,200), data=data)
plt.show()
```

```python
# Using scatterplot
sns.scatterplot(x='total_bill', y='tip', hue='smoker',
            style='sex', size='size', data=data)
plt.show()
```
