#### __`Violin plot`__
- Violin plots are method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot(kde).
- Advantage of violin plot over box plot is that beside form showing a five number summary is also shows the entire distribution of the data.
- Violin plots are often used to compare the distribution of a given variable across some categories.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sns.violinplot(tips['total_bill'])
plt.show()
```

```python
sns.catplot(x='day',y='total_bill', kind='violin',data=tips)
plt.show()
```

```python
# using 'hue' sematic parameter
sns.catplot(x='day', y='total_bill', kind='violin',hue='sex',split='True', data=tips)
plt.show()
```
