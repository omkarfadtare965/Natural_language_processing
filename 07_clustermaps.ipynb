#### __`cluster maps`__
- It is only a heatmap but the only difference is that it's clustered. 
- Forms clusters of similar rows and similar columns together.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
flights = sns.load_dataset('flights')
flights.head(5)
```

```python
# clustering happens because of euclidean metrics parameters; by finding the euclidean distance between two points.
x = flights.pivot_table(index='year', columns='month',
                        values='passengers', aggfunc='sum')
sns.clustermap(x)
plt.show()
```

```python
# changing many parameters
x = flights.pivot_table(index='year', columns='month',
                        values='passengers', aggfunc='sum')
sns.clustermap(x,metric='correlation',z_score=0,cmap='viridis', annot=True)
plt.show()
```

```python
# clustering on the basis of columns only
x = flights.pivot_table(index='year', columns='month',
                        values='passengers', aggfunc='sum')
sns.clustermap(x,row_cluster=False, metric='correlation', z_score=0, cmap='viridis', annot=True)
plt.show()
```
