#### __`Bar plot & Count plot`__
- Rather than giving you the information about distribution of the data they give you the information about central tendency.
- Information of numeric data with respect to categorical data.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

#### __`Bar plot`__

```python
sns.catplot(x='smoker',y='total_bill', kind='bar', data=tips)
plt.show()
# these black bars are known as error bars, they are showing the confidence interval to calculate mean for the population.
# basically, these plots are trying to infer the value of mean for the population with the confidence interval of 95%.
# default behaviour of bar plot is to calculate the mean of the numeric value but you can tweak this by using 'estimator' parameter.
```

```python
# using 'hue' semantic parameter 
sns.catplot(x='smoker', y='total_bill', hue='sex', kind='bar', data=tips)
plt.show()
```

```python
# using 'estimator' parameter # by default value is estimator= mean
sns.catplot(x='smoker', y='total_bill', hue='sex', estimator=np.median, kind='bar', data=tips)
plt.show()
```

```python
# changing the estimator parameter value to variance
sns.catplot(x='smoker', y='total_bill', hue='sex',
            estimator=np.var, kind='bar', data=tips)
plt.show()
```

```python
# changing the estimator parameter value to standard deviation
sns.catplot(x='smoker', y='total_bill', hue='sex',
            estimator=np.std, kind='bar', data=tips)
plt.show()
```

#### __`Count plot`__
count plots are the special cases of bar plots where you actually do not consider any numeric value, you just take a categorical value and find the count of that value. It works same as histogram.

```python
sns.catplot(x='sex',hue='smoker', kind='count', data=tips)
plt.show()
```
