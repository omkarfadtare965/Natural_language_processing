### __`Dist plots`__
Distplots are used to do univariate analysis on a continuous/numerical data. Dist plots combines three things in one graph.
- Histogram
- Kernel density estimation(KDE)
- Rug plot

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
titanic=sns.load_dataset('titanic')
titanic.head(5)
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
# it is a combination of two different graphs kde plot(to plot probability density function) and histogram(distribution of numeric data)
sns.distplot(titanic['age'])
plt.show()
```

```python
# Plotting only histogram
sns.distplot(titanic['age'],kde=False)
plt.show()
```

```python
# increasing or decreasing bins
sns.distplot(titanic['age'], kde=False, bins=10)
plt.show()
```

```python
# increasing or decreasing bins
sns.distplot(titanic['age'], kde=False, bins=100)
plt.show()
```

```python
# Plotting only kde
sns.distplot(titanic['age'], hist=False)
plt.show()
```

```python
# Plotting only rug plot
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
plt.show()
```

```python
# ages of people who survived 
sns.displot(titanic[titanic['survived']==1]['age'],kde=True)
sns.displot(titanic[titanic['survived'] == 0]['age'],kde=True)
plt.show()
```
