#### __`Heatmap`__
- A heat map is a graphical representation of data where the individual values contained in matrix are represented as colors.
- Generally, drawn between two categorical data or between discrete numerical data.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
flights = sns.load_dataset('flights')
flights.head(5)
```

```python
x = flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
sns.heatmap(x)
plt.show()
```

```python
x = flights.pivot_table(index='year', columns='month',
                        values='passengers', aggfunc='sum')
sns.heatmap(x,linewidths=0.5, annot=True, fmt='d')
plt.show()
```

```python
# changing the color and other parameters
x = flights.pivot_table(index='year', columns='month',
                        values='passengers', aggfunc='sum')
sns.heatmap(x, linewidths=0.5, annot=True, fmt='d',cmap='summer')
plt.show()
```

```python
# changing the color 
x = flights.pivot_table(index='year', columns='month',
                        values='passengers', aggfunc='sum')
sns.heatmap(x, linewidths=0.5, annot=True, fmt='d',cmap='viridis')
plt.show()
```
