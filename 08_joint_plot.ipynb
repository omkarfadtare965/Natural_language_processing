### __`Joint plot`__
- Joint plot create a multi-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.
- We can use joint plot in the following 4 manners
  - Scatter plot(bivariate): in addition it plots histogram on both the axis(univariate)
  - Hexbin plots: 
  - Kde plot
  - Reg plot

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

#### __`Scatter plot`__

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

#### __`Hexbins`__

```python
# it is nothing but a 2D histogram
sns.jointplot(x='total_bill', y='tip',kind='hex', data=tips)
plt.show()
```

#### __`kde plot`__

```python
# middle plot is know as contour
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)
plt.show()
```

#### __`Regression plot`__

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
plt.show()
```
